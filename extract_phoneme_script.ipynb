Static appraoch . Sentecne output that is phonemes should be present in ratios.jsson file. thn it will work. Limitation.

```python
#extract phonemes either with g2p or epitran
def get_g2p_phoneme(word):
  g2p = G2p()
  return g2p(word)

#either get it with epitran(multi lang code)
def get_epitran_phoneme(word):
  epi = epitran.Epitran('eng-Latn')
  return epi.transliterate(word)


#calculate ratios from ratio.json
def extract_phoneme_ratios(phonemes):
  durations = [] #store the duration between current n next phoneme
  for i in range(len(phonemes)-1):
    current = phonemes[i]
    next_phoneme = phonemes[i+1]
    ratio = phoneme_ratios.get(current,{}).get(next_phoneme,{}).get('-',0)
    durations.append(ratio)

  return durations

word = "hello i am abhishek"
phonemes = get_g2p_phoneme(word)  # or use get_phonemes_epitran(word) if using Epitran
durations = extract_phoneme_ratios(phonemes)

print(f"Phonemes for '{word}': {phonemes}")
print(f"Phoneme durations: {durations}")
```
